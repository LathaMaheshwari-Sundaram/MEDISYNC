```python
# Install Gradio
!pip install -U gradio

import gradio as gr
import pandas as pd
import random
from datetime import datetime

# In-memory billing records
billing_records = []

# 📌 Function: Make payment
def make_payment(patient_number, amount):
    global billing_records

    if not str(patient_number).strip() or amount <= 0:
        return "⚠️ Enter valid patient number and amount!", pd.DataFrame(columns=["Patient Number","Amount","Status","Date & Time"])

    # Simulate random payment success/failure
    status = random.choice(["✅ Payment Successful", "❌ Payment Failed"])

    new_entry = {
        "Patient Number": patient_number,
        "Amount": f"₹{amount}",
        "Status": status,
        "Date & Time": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    }
    billing_records.append(new_entry)

    df = pd.DataFrame(billing_records)
    return status, df

# 📌 Function: View payment history
def view_payments():
    if len(billing_records) == 0:
        return "⚠️ No payments yet!", pd.DataFrame(columns=["Patient Number","Amount","Status","Date & Time"])
    df = pd.DataFrame(billing_records)
    return "📑 Billing History", df

# ---------------- UI ----------------
with gr.Blocks() as demo:
    gr.Markdown("# 💳 Patient Billing Dashboard")

    with gr.Tab("💰 Make Payment"):
        with gr.Row():
            patient_number = gr.Textbox(label="Enter Patient Number / Mobile Number")
            amount = gr.Number(label="Enter Amount (₹)", value=500)
        pay_btn = gr.Button("💳 Pay Now")
        pay_msg = gr.Textbox(label="Payment Status")
        pay_table = gr.Dataframe(headers=["Patient Number", "Amount", "Status", "Date & Time"], wrap=True)

        pay_btn.click(
            fn=make_payment,
            inputs=[patient_number, amount],
            outputs=[pay_msg, pay_table]
        )

    with gr.Tab("📑 View Billing History"):
        view_btn = gr.Button("🔄 Refresh Billing History")
        view_msg = gr.Textbox(label="Status")
        view_table = gr.Dataframe(headers=["Patient Number", "Amount", "Status", "Date & Time"], wrap=True)

        view_btn.click(fn=view_payments, outputs=[view_msg, view_table])

# Launch app
demo.launch(share=True)
```
