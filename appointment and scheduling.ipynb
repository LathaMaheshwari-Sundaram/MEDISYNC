```python
# Install Gradio
!pip install gradio

import gradio as gr
import pandas as pd

# In-memory appointment database
appointments = []

# Dropdown options
hospitals = ["Apollo Hospital", "Fortis Healthcare", "AIIMS", "CMC Vellore"]
doctors = ["Dr. Arjun (Cardiologist)", "Dr. Meera (Dermatologist)", "Dr. Rahul (Neurologist)", "Dr. Anita (Pediatrician)"]
dates = ["2025-09-01", "2025-09-02", "2025-09-03", "2025-09-04"]
times = ["09:00 AM", "10:30 AM", "12:00 PM", "02:00 PM", "04:00 PM"]

# Function to book an appointment + show patient history
def book_appointment(patient, hospital, doctor, date, time):
    global appointments

    if not patient.strip():
        return "⚠️ Please enter patient name!", pd.DataFrame(), pd.DataFrame()

    # New appointment entry
    new_entry = {
        "Patient Name": patient,
        "Hospital": hospital,
        "Doctor": doctor,
        "Date": date,
        "Time": time
    }
    appointments.append(new_entry)
    df = pd.DataFrame(appointments)

    # Filter only this patient's past appointments
    patient_history = df[df["Patient Name"].str.lower() == patient.lower()]

    return (f"✅ Appointment booked for {patient} at {hospital} with {doctor} on {date} at {time}",
            df,
            patient_history)

# Function to view all appointments
def view_appointments():
    if len(appointments) == 0:
        return "⚠️ No appointments yet!", pd.DataFrame(columns=["Patient Name","Hospital","Doctor","Date","Time"])
    df = pd.DataFrame(appointments)
    return "📅 Current Appointments", df

# Build Gradio UI
with gr.Blocks() as demo:
    gr.Markdown("# 🏥 Appointment Scheduling System")

    with gr.Tab("📌 Book Appointment"):
        with gr.Row():
            patient = gr.Textbox(label="Enter Patient Name", placeholder="Type patient name here...")
            hospital = gr.Dropdown(hospitals, label="Select Hospital")
        with gr.Row():
            doctor = gr.Dropdown(doctors, label="Select Doctor")
        with gr.Row():
            date = gr.Dropdown(dates, label="Select Date")
            time = gr.Dropdown(times, label="Select Time")

        book_btn = gr.Button("✅ Book Appointment")
        book_msg = gr.Textbox(label="Booking Status")

        gr.Markdown("### 📅 All Appointments")
        book_table = gr.Dataframe(headers=["Patient Name", "Hospital", "Doctor", "Date", "Time"], wrap=True)

        gr.Markdown("### 📜 Patient's Previous Appointments")
        history_table = gr.Dataframe(headers=["Patient Name", "Hospital", "Doctor", "Date", "Time"], wrap=True)

        book_btn.click(
            fn=book_appointment,
            inputs=[patient, hospital, doctor, date, time],
            outputs=[book_msg, book_table, history_table]
        )

    with gr.Tab("📂 View All Appointments"):
        view_btn = gr.Button("🔄 Refresh Appointments")
        view_msg = gr.Textbox(label="Status")
        view_table = gr.Dataframe(headers=["Patient Name", "Hospital", "Doctor", "Date", "Time"], wrap=True)

        view_btn.click(fn=view_appointments, outputs=[view_msg, view_table])

# Launch app
demo.launch(share=True)
```
