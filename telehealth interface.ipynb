```python
import gradio as gr

# In-memory user storage
users = {}
appointments = {}

# Sample doctor data
doctors = {
    "D001": {
        "name": "Dr. Smith",
        "specialization": "Cardiologist",
        "slots": ["10:00 AM", "11:00 AM", "2:00 PM"]
    },
    "D002": {
        "name": "Dr. Alice",
        "specialization": "Dermatologist",
        "slots": ["9:00 AM", "1:00 PM", "4:00 PM"]
    },
    "D003": {
        "name": "Dr. John",
        "specialization": "Pediatrician",
        "slots": ["10:30 AM", "12:30 PM", "3:00 PM"]
    }
}

# Global session variable (simulated)
session = {"user_id": None}

def register(name, age, phone, user_id, password):
    if user_id in users:
        return "❌ User ID already exists."
    users[user_id] = {
        "name": name,
        "age": age,
        "phone": phone,
        "password": password
    }
    return f"✅ Registered successfully as {name}!"

def login(user_id, password):
    if user_id in users and users[user_id]["password"] == password:
        session["user_id"] = user_id
        return f"✅ Logged in as {users[user_id]['name']}!"
    return "❌ Invalid User ID or password."

def list_doctors():
    output = ""
    for doc_id, info in doctors.items():
        output += f"{doc_id}: {info['name']} ({info['specialization']})\nSlots: {', '.join(info['slots'])}\n\n"
    return output if output else "No doctors available."

def book_appointment(doc_id, slot):
    user_id = session["user_id"]
    if not user_id:
        return "❌ Please log in first."

    if doc_id not in doctors:
        return "❌ Invalid Doctor ID."

    if slot not in doctors[doc_id]["slots"]:
        return "❌ Slot not available."

    # Remove the slot
    doctors[doc_id]["slots"].remove(slot)

    # Save the appointment
    appointments[user_id] = {
        "doctor": doctors[doc_id]["name"],
        "specialization": doctors[doc_id]["specialization"],
        "slot": slot,
        "link": "https://meet.google.com/xyz-abc-def"
    }

    return f"✅ Appointment confirmed with {doctors[doc_id]['name']} at {slot}.\n🔗 Meet Link: https://meet.google.com/xyz-abc-def"

def view_summary():
    user_id = session["user_id"]
    if not user_id:
        return "❌ Please log in first."

    if user_id in appointments:
        appt = appointments[user_id]
        return (
            f"📋 Appointment Summary:\n\n"
            f"Doctor: {appt['doctor']} ({appt['specialization']})\n"
            f"Slot: {appt['slot']}\n"
            f"Link: {appt['link']}"
        )
    else:
        return "You have no appointments yet."

def logout():
    session["user_id"] = None
    return "👋 Logged out."

# Gradio Interface

with gr.Blocks() as demo:
    gr.Markdown("## 🏥 Telehealth Appointment Booking System")

    with gr.Tab("📝 Register"):
        name = gr.Textbox(label="Full Name")
        age = gr.Textbox(label="Age")
        phone = gr.Textbox(label="Phone")
        user_id_reg = gr.Textbox(label="Create User ID")
        password_reg = gr.Textbox(label="Password", type="password")
        register_btn = gr.Button("Register")
        register_output = gr.Textbox(label="Output")
        register_btn.click(register, inputs=[name, age, phone, user_id_reg, password_reg], outputs=register_output)

    with gr.Tab("🔐 Login"):
        user_id_login = gr.Textbox(label="User ID")
        password_login = gr.Textbox(label="Password", type="password")
        login_btn = gr.Button("Login")
        login_output = gr.Textbox(label="Output")
        login_btn.click(login, inputs=[user_id_login, password_login], outputs=login_output)

    with gr.Tab("👨‍⚕️ View Doctors"):
        view_doctors_btn = gr.Button("Show Available Doctors")
        doctor_output = gr.Textbox(label="Doctors")
        view_doctors_btn.click(list_doctors, outputs=doctor_output)

    with gr.Tab("📅 Book Appointment"):
        doc_id_input = gr.Textbox(label="Enter Doctor ID (e.g., D001)")
        slot_input = gr.Textbox(label="Enter Time Slot (e.g., 10:00 AM)")
        book_btn = gr.Button("Book Now")
        book_output = gr.Textbox(label="Booking Result")
        book_btn.click(book_appointment, inputs=[doc_id_input, slot_input], outputs=book_output)

    with gr.Tab("📋 Appointment Summary"):
        summary_btn = gr.Button("Show Summary")
        summary_output = gr.Textbox(label="Summary")
        summary_btn.click(view_summary, outputs=summary_output)

    with gr.Tab("🚪 Logout"):
        logout_btn = gr.Button("Logout")
        logout_output = gr.Textbox(label="Output")
        logout_btn.click(logout, outputs=logout_output)

demo.launch()
```
