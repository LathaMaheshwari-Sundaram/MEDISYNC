```python
# Install latest Gradio
!pip install -U gradio

import gradio as gr
import pandas as pd
from datetime import datetime, timedelta

# In-memory storage for messages/notifications
notifications = [
    {"Patient": "Rahul", "Message": "⏰ Appointment tomorrow with Dr. Arjun at Apollo Hospital, 10:30 AM", "Type": "Appointment"},
    {"Patient": "Meera", "Message": "💊 Reminder: Please refill your medicine 'Metformin' before 2025-09-05", "Type": "Medicine Refill"}
]

# 📌 Function: View notifications
def view_notifications():
    if len(notifications) == 0:
        return "⚠️ No notifications yet!", pd.DataFrame(columns=["Patient","Message","Type"])
    df = pd.DataFrame(notifications)
    return "📩 Notifications Dashboard", df

# 📌 Function: Add medicine refill reminder
def add_refill_reminder(patient, medicine, days):
    global notifications
    if not patient.strip() or not medicine.strip():
        return "⚠️ Enter patient name and medicine!", pd.DataFrame()

    due_date = datetime.today() + timedelta(days=int(days))
    message = f"💊 Reminder: {patient}, please refill your medicine **{medicine}** before {due_date.date()}."
    notifications.append({"Patient": patient, "Message": message, "Type": "Medicine Refill"})

    df = pd.DataFrame(notifications)
    return "✅ Medicine reminder added!", df

# ---------------- UI ----------------
with gr.Blocks() as demo:
    gr.Markdown("# 🔔 Patient Notifications Dashboard")

    with gr.Tab("📩 View Messages & Reminders"):
        notif_btn = gr.Button("🔄 Refresh Notifications")
        notif_msg = gr.Textbox(label="Status")
        notif_table = gr.Dataframe(headers=["Patient", "Message", "Type"], wrap=True)
        notif_btn.click(fn=view_notifications, outputs=[notif_msg, notif_table])

    with gr.Tab("💊 Add Medicine Reminder"):
        with gr.Row():
            r_patient = gr.Textbox(label="Patient Name")
            medicine = gr.Textbox(label="Medicine Name")
            days = gr.Number(label="Days Until Reminder", value=7)
        refill_btn = gr.Button("➕ Add Reminder")
        refill_msg = gr.Textbox(label="Status")
        refill_table = gr.Dataframe(headers=["Patient", "Message", "Type"], wrap=True)

        refill_btn.click(
            fn=add_refill_reminder,
            inputs=[r_patient, medicine, days],
            outputs=[refill_msg, refill_table]
        )

# Launch app
demo.launch(share=True)
```
