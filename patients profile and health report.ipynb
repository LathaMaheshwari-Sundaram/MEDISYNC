```python
# Install Gradio
!pip install gradio

import gradio as gr

# Sample patient data
patient_data = {
    "Name": "John Doe",
    "Age": "45",
    "Gender": "Male",
    "Blood Group": "O+",
    "Mobile": "+91 9876543210",
    "Email": "johndoe@email.com",
    "Address": "123 Main Street, Chennai",
    "Allergies": "Penicillin",
    "Chronic Conditions": "Diabetes, Hypertension",
    "Current Medications": "Metformin, Aspirin",
    "Emergency Contact": "+91 9123456789 (Wife)",
    "Health Records": {
        "Blood Test Report (12 Aug 2025)": "Normal",
        "ECG Scan (10 Aug 2025)": "Minor irregularity detected",
        "Prescription (Dr. Smith)": "Continue Metformin, Aspirin"
    }
}

# Function to display patient profile
def show_profile():
    profile = f"""
    👤 **Basic Information**
    - Name: {patient_data['Name']}
    - Age: {patient_data['Age']}
    - Gender: {patient_data['Gender']}
    - Blood Group: {patient_data['Blood Group']}
    - Mobile: {patient_data['Mobile']}
    - Email: {patient_data['Email']}
    - Address: {patient_data['Address']}

    💊 **Medical Information**
    - Allergies: {patient_data['Allergies']}
    - Chronic Conditions: {patient_data['Chronic Conditions']}
    - Current Medications: {patient_data['Current Medications']}
    - Emergency Contact: {patient_data['Emergency Contact']}

    📂 **Health Records**
    """
    for record, detail in patient_data["Health Records"].items():
        profile += f"\n    - {record}: {detail}"
    return profile

# Create Gradio Interface
with gr.Blocks() as demo:
    gr.Markdown("# 🏥 Patient Profile Dashboard")
    btn = gr.Button("Show Patient Profile")
    output = gr.Textbox(label="Patient Details", lines=20)
    btn.click(fn=show_profile, outputs=output)

# Launch the app
demo.launch(share=True)
```
